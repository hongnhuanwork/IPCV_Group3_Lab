```python
import cv2
import mediapipe as mp
import matplotlib.pyplot as plt

# Khởi tạo công cụ vẽ và giải pháp nhận diện
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils

# Cấu hình mô hình
# model_selection=0: phù hợp cho khuôn mặt cách camera trong vòng 2 mét.
# min_detection_confidence: ngưỡng tin cậy tối thiểu để chấp nhận một dự đoán (Precision)[cite: 809].
face_detection = mp_face_detection.FaceDetection(model_selection=0, min_detection_confidence=0.5)
```

```python
def detect_faces(image_path):
    # Bước 1: Nhận hình ảnh đầu vào (Input Image) [cite: 77, 131]
    image = cv2.imread(image_path)
    if image is None:
        return "Không tìm thấy ảnh"
    
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Bước 2: Trích xuất đặc trưng và dự đoán (Inference) [cite: 22, 82]
    results = face_detection.process(image_rgb)
    
    # Bước 3: Giải mã và hiển thị kết quả (Decoding & Visualization) [cite: 32, 113]
    if results.detections:
        print(f"Tìm thấy {len(results.detections)} khuôn mặt.")
        for detection in results.detections:
            # Vẽ khung bao (Bounding Box) và các điểm mốc
            mp_drawing.draw_detection(image, detection)
            
            # Trích xuất độ tự tin (Score) [cite: 173]
            score = detection.score[0]
            print(f"Độ tự tin: {score:.2f}")
    else:
        print("Không tìm thấy khuôn mặt nào.")
        
    # Hiển thị ảnh kết quả bằng Matplotlib (phù hợp với Jupyter)
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

# Thử nghiệm với một file ảnh (Hãy thay đổi đường dẫn file của bạn)
# detect_faces('path_to_your_image.jpg')
```
