```python
import cv2
```

```python
img = cv2.imread("lab01.jpg")

cv2.imshow("Image",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite("lab01_copy.png", img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow("HSV Image", hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2.imshow("LAB Image", lab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
crop = img[100:300, 200:400]

cv2.imshow("Crop Image", crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize = cv2.resize(img, (400, 300))
cv2.imshow("Resize Image", resize)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
resize = cv2.resize(img, None, fx=0.5, fy=0.5)
cv2.imshow("Resize 50%", resize)
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
cv2.line(resize, (50,50), (300,50), (0,255,0), 3)
cv2.rectangle(resize, (50,100), (300,200), (255,0,0), 2)
cv2.circle(resize, (200,200), 50, (0,0,255), 3)

cv2.putText(resize,
            "Group 3",
            (50,50),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (255,255,255),
            2)

cv2.imshow("Image with Text", resize)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
