```python
import cv2


# Đọc ảnh jpg
img = cv2.imread("lab01.jpg")
```

# Câu 2

```python

# Hiển thị ảnh
cv2.imshow("Image", img)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

```python
# Lưu sang định dạng khác (ví dụ PNG)
cv2.imwrite("image.png", img)
```

# Câu 3

```python

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# hiển thị Gray
cv2.imshow("Gray", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
hsv  = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# hiển thị HSV
cv2.imshow("HSV", hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lab  = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

# hiển thị LAB
cv2.imshow("LAB", lab)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Câu 4

```python
# cắt vùng ảnh (y1:y2, x1:x2)
crop = img[100:300, 200:400]


cv2.imshow("Crop", crop)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python

small = cv2.resize(img, None, fx=0.3, fy=0.3)

cv2.imshow("Resize", small)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Câu 5

```python
small = cv2.resize(img, None, fx=0.3, fy=0.3)

# vẽ đường thẳng
cv2.line(small, (20, 20), (200, 20), (0, 255, 0), 2)

# vẽ hình chữ nhật
cv2.rectangle(small, (50, 50), (200, 150), (255, 0, 0), 2)

# vẽ hình tròn
cv2.circle(small, (300, 150), 40, (0, 0, 255), 2)

# thêm văn bản
cv2.putText(small,
            "Nguyen Minh Thinh",
            (20, 250),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (255, 255, 0),
            2)

# hiển thị
cv2.imshow("Result", small)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
