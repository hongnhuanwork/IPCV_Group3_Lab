```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Đọc ảnh
img = cv2.imread("lab01.jpg")

# Chuyển BGR -> RGB để hiển thị đúng màu
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Hiển thị ảnh
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
```

```python
# Lưu ảnh thành PNG
cv2.imwrite("output.png", img)

print("Image saved as output.png")
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(hsv)
plt.title("HSV Image")
plt.axis("off")
```

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)

plt.imshow(lab)
plt.title("LAB Image")
plt.axis("off")
```

```python
crop = img_rgb[100:400, 100:400]

plt.imshow(crop)
plt.title("Cropped Image")
plt.axis("off")
```

```python
resize1 = cv2.resize(img_rgb, (300,300))

plt.imshow(resize1)
plt.title("Resize 300x300")
plt.axis("off")
```

```python
resize2 = cv2.resize(img_rgb, None, fx=0.5, fy=0.5)

plt.imshow(resize2)
plt.title("Resize 50%")
plt.axis("off")
```

```python
draw = img_rgb.copy()

# Vẽ đường thẳng
cv2.line(draw, (50,50), (400,50), (255,0,0), 3)

# Vẽ hình chữ nhật
cv2.rectangle(draw, (100,100), (300,300), (0,255,0), 3)

# Vẽ hình tròn
cv2.circle(draw, (500,300), 80, (0,0,255), 3)

plt.imshow(draw)
plt.title("Shapes")
plt.axis("off")
```

```python
cv2.putText(draw,
            "Hong Nhuan",
            (50,450),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (255,255,0),
            2)

plt.imshow(draw)
plt.title("Text on Image")
plt.axis("off")
```
