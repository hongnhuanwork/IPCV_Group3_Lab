```python
import cv2
import matplotlib.pyplot as plt
```

<h3> Hiển thị ảnh <h3>

```python
# Đọc ảnh (thay 'image.jpg' bằng tên file của bạn)
img = cv2.imread("lab01.jpg")
```

```python
# Hiển thị ảnh
cv2.imshow('Original Image', img)
```

```python
# Lưu ảnh với định dạng khác (ví dụ .png)
cv2.imwrite('output_image.png', img)
```

```python
cv2.waitKey(0) # Đợi nhấn phím bất kỳ để đóng cửa sổ
cv2.destroyAllWindows()
```

```python
img_rgb       = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # Để hiển thị chuẩn
img_gray      = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Grayscale
img_hsv       = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)   # HSV
img_lab       = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)   # LAB
```

```python
# 3. Hiển thị tất cả bằng Matplotlib để tránh crash Kernel
plt.figure(figsize=(15, 10))
```

```python
# Ảnh Gốc (RGB)
plt.subplot(2, 2, 1)
plt.imshow(img_rgb)
plt.title('Original (RGB)')
plt.axis('off')
```

```python
# Ảnh Xám (Grayscale)
plt.subplot(2, 2, 2)
plt.imshow(img_gray, cmap='gray') # Lưu ý phải có cmap='gray'
plt.title('Grayscale')
plt.axis('off')
```

```python
# Ảnh HSV
plt.subplot(2, 2, 3)
plt.imshow(img_hsv)
plt.title('HSV Space')
plt.axis('off')
```

```python
# Ảnh LAB
plt.subplot(2, 2, 4)
plt.imshow(img_lab)
plt.title('LAB Space')
plt.axis('off')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off') # Ẩn trục tọa độ
plt.show()
```

```python
# 4.1. Cắt một vùng hình ảnh bất kỳ bằng NumPy Slicing
# Cú pháp: img[y_start:y_end, x_start:x_end]
y_start, y_end = 100, 400
x_start, x_end = 200, 600
cropped_img = img_rgb[y_start:y_end, x_start:x_end]
```

```python
# Hiển thị kết quả
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1); plt.imshow(cropped_img); plt.title('Vùng ảnh đã cắt')
```

```python
# 4.2. Thay đổi kích thước (Resize)
# Thay đổi theo kích thước cố định (Ví dụ: 300x300)
resized_fixed = cv2.resize(img_rgb, (300, 300))
```

```python
# Thay đổi theo tỷ lệ (Ví dụ: giảm còn 50% kích thước gốc)
resized_ratio = cv2.resize(img_rgb, None, fx=0.5, fy=0.5)
```

```python
plt.subplot(1, 2, 2); plt.imshow(resized_fixed); plt.title('Ảnh Resize 300x300')
```

```python
# Tạo bản sao để vẽ
drawing_img = img_rgb.copy()
```

```python
# 5.1. Vẽ các hình cơ bản (Sử dụng OpenCV nhưng hiển thị bằng Matplotlib)
# Vẽ đường thẳng (Start, End, Màu RGB, Độ dày)
cv2.line(drawing_img, (50, 50), (300, 50), (255, 0, 0), 5)
```

```python
# Vẽ hình chữ nhật (Top-Left, Bottom-Right, Màu RGB, Độ dày)
cv2.rectangle(drawing_img, (50, 100), (250, 300), (0, 255, 0), 3)
```

```python
# Vẽ hình tròn (Tâm, Bán kính, Màu RGB, Độ dày -1 để tô kín)
cv2.circle(drawing_img, (500, 200), 70, (0, 0, 255), -1)
```

```python
# 5.2. Thêm văn bản (Text, Tọa độ, Font, Scale, Màu, Độ dày)
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(drawing_img, 'DINO Haz - OpenCV 101', (50, 450), font, 1.2, (255, 255, 255), 3)
```

```python
# Hiển thị kết quả cuối cùng
plt.figure(figsize=(8, 8))
plt.imshow(drawing_img)
plt.title('Kết quả vẽ hình và thêm văn bản')
plt.axis('off')
plt.show()
```
