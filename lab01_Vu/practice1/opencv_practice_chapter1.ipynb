## 2. Đọc và hiển thị ảnh

```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('test.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')
```

## 3. Chuyển đổi không gian màu

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.axis('off')
```

## 4. Crop và Resize

```python
crop = image[100:400,200:500]
resize = cv2.resize(image,(300,300))

plt.imshow(cv2.cvtColor(resize, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')
```

## 5. Vẽ hình cơ bản

```python
draw = image.copy()

h, w, _ = draw.shape

center_x = w // 2
center_y = h // 2

thick = 20

cv2.line(draw,(center_x-200,center_y-200),(center_x+200,center_y+200),(0,255,0),thick)

cv2.rectangle(draw,(center_x-200,center_y-150),(center_x+200,center_y+150),(255,0,0),thick)

cv2.circle(draw,(center_x,center_y),120,(0,0,255),thick)

plt.figure(figsize=(8,6))
plt.imshow(cv2.cvtColor(draw, cv2.COLOR_BGR2RGB))
plt.title('Shapes')
plt.axis('off')
```

## 6. Thêm văn bản vào ảnh

```python
text_img = image.copy()

text = "Hello OpenCV"
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 8    
thickness = 4      

(text_width, text_height), _ = cv2.getTextSize(text, font, font_scale, thickness)

img_height, img_width, _ = text_img.shape

x = (img_width - text_width) // 2
y = (img_height + text_height) // 2

cv2.putText(text_img, text, (x, y), font, font_scale, (0,0,255), thickness+16)

cv2.putText(text_img, text, (x, y), font, font_scale, (0,0,255), thickness)

plt.imshow(cv2.cvtColor(text_img, cv2.COLOR_BGR2RGB))
plt.title("Text on Image")
plt.axis("off")
```
